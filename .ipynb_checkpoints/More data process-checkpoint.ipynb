```python
import pandas as pd
import numpy as np
```

```python
GDP = pd.read_csv("./data/GDP.csv")
GDP.head(3)
```

```python
id_vars = ['countrycode', 'countryname', 'region', 'regionname', 'adminregion', 'adminregionname', 
           'incomelevel', 'incomelevelname', 'lendingtype', 'lendingtypename', 'indicatorname', 'indicatorcode']
```

```python
GDP = pd.melt(GDP, id_vars=id_vars, var_name='year_indicator', value_name='indicator_value')
GDP.head(3)
```

```python
GDP['year_indicator'] = GDP['year_indicator'].str[2:]
GDP['year_indicator'] = GDP['year_indicator'].astype(int)
```

```python
GDP = GDP[(GDP['year_indicator'] >= 1992) & (GDP['year_indicator'] <= 2020)]
GDP.head(10)
```

```python
GDP.drop(['indicatorname'], axis=1, inplace=True)
GDP.head(10)
```

```python
GDP = GDP.rename(columns={'year_indicator': 'year'})
GDP.head(1)
```

```python
GDP = GDP.rename(columns={'indicator_value': 'GDP(current $USD)'})
GDP.head(1)
```

```python
missing_values = GDP.isna()
missing_counts = missing_values.sum()
print(missing_counts)
```

```python
GDP['GDP(current $USD)'] = GDP['GDP(current $USD)'].interpolate(method='linear')
missing_values = GDP.isna()
missing_counts = missing_values.sum()
print(missing_counts)
```

```python
GDP.to_csv('./data/Purified_GDP.csv', index=False, sep=';', quotechar='"', encoding='utf-8')
```

```python
gshockg = pd.merge(gshockt, GDP, how='left', on=['countryname', 'year'])
gshockg.head(10)
```

```python
gshockgg = pd.merge(gshockg, gs, on=['year', 'countryname'])
gshockgg.head(10)
```

```python
missing_values = gshockgg.isna()
missing_counts = missing_values.sum()

print(missing_counts)
```

```python
g_list = []
```

```python
for i in range(1991, 2021):
    temp = gshockgg[gshockgg['year'] == i]
    sigma = 0
    cn_exr = temp[temp['countryname'] == 'China']['exchange rate (to $USD)']
    for index, row in temp.iterrows():
        if row['countryname'] == 'China':
            cn_exr = row['exchange rate (to $USD)']
        if row.isna().any():
            pass
        else:
            sigma = sigma + row['import (thousand $USD)'] * 1000 / row['general government final consumption expenditure (Current $USD)']* (row['import (thousand $USD)'] * 1000 + row['export (thousand $USD)']) / row['GDP(current $USD)'] * 1000 * row['general government final consumption expenditure (Current $USD)']* row['exchange rate (to $USD)']
    res = sigma / cn_exr
    g_list.append(res)
```
