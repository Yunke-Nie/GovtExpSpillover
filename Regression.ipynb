```python
data = pre_gs.set_index('year')
data = data.dropna()
```

```python
data['deltaGDP'] = data.groupby('countryname')['GDP(current $USD)'].diff()
data
```

```python
data['deltaGovtExp'] = data.groupby('countryname')['general government final consumption expenditure (Current $USD)'].diff()

data
```

```python
data['GShock'] = data['import (thousand $USD)'] / data['general government final consumption expenditure (Current $USD)'] * (data['import (thousand $USD)'] + data['export (thousand $USD)']) / data['GDP(current $USD)'] * 1000 * 1000 * data['general government final consumption expenditure (Current $USD)'].shift(1) 
data
```

```python
data = data.dropna()
```

```python
deltaYGDP_lag1 = data['deltaGDP'].shift(1) / data['GDP(current $USD)'].shift(1)
deltaYGDP_lag2 = data['deltaGDP'].shift(2) / data['GDP(current $USD)'].shift(2)
deltaYGDP_lag3 = data['deltaGDP'].shift(3) / data['GDP(current $USD)'].shift(3)
GovtExp_lag1 = data['deltaGovtExp'].shift(1)
GovtExp_lag2 = data['deltaGovtExp'].shift(2)
GovtExp_lag3 = data['deltaGovtExp'].shift(3)
```
