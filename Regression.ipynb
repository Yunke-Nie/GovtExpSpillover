```python
import pandas as pd
import numpy as np
```

```python
pre_gs = pd.read_csv("./data/purified/gshock.csv")
pre_gs.head(30)
```

```python
ref_cn = pre_gs[pre_gs['countryname'] == 'China']
ref_cn
```

```python
ref_cn.reindex(ref_cn['year'])
ref_cn
```

```python
ref_cn['exchange rate (to $USD)'] = np.log(ref_cn['exchange rate (to $USD)'])
ref_cn['GDP(current $USD)'] = np.log(ref_cn['GDP(current $USD)'])
ref_cn['general government final consumption expenditure (Current $USD)'] = np.log10(ref_cn['general government final consumption expenditure (Current $USD)'])
ref_cn
```

```python
pre_gs = pre_gs.dropna()
```

```python
pre_gs['GShock'] = ''
pre_gs.head(3)
```

```python
pre_gs['exchange rate (to $USD)'] = np.log(pre_gs['exchange rate (to $USD)'])
pre_gs['import (thousand $USD)'] = np.log(pre_gs['import (thousand $USD)'] * 1000)
pre_gs['export (thousand $USD)'] = np.log(pre_gs['export (thousand $USD)'] * 1000)
pre_gs['GDP(current $USD)'] = np.log(pre_gs['GDP(current $USD)'])
pre_gs['general government final consumption expenditure (Current $USD)'] = np.log(pre_gs['general government final consumption expenditure (Current $USD)'])
```

```python
pre_gs.head(30)
```

```python
new_gs = []
for i in range(1992, 2021):
    cur = pre_gs[pre_gs['year'] == i]
    lastyear = pre_gs[pre_gs['year'] == i - 1]
    for index, row in cur.iterrows():
        if i >= 1993:
            new_gs.append(row['import (thousand $USD)'] 
                         / ref_cn.loc[ref_cn['year'] == i, 'general government final consumption expenditure (Current $USD)'].iloc[0]
                         * (row['export (thousand $USD)'] + row['import (thousand $USD)'])
                         / row['GDP(current $USD)'] * ref_cn.loc[ref_cn['year'] == i - 1, 'general government final consumption expenditure (Current $USD)'].iloc[0])
        else:
            new_gs.append(0)
    
```

```python
new_gs
```

```python
pre_gs['GShock'] = new_gs
pre_gs
```

```python
data = pre_gs.set_index('year')
data = data.dropna()
```

```python
data
```

```python
data['deltaGDP'] = data.groupby('countryname')['GDP(current $USD)'].diff()
data
```

```python
data['deltaGovtExp'] = data.groupby('countryname')['general government final consumption expenditure (Current $USD)'].diff()

data
```

```python
data = data.dropna()
```

```python
deltaYGDP_lag1 = data['deltaGDP'].shift(1) / data['GDP(current $USD)'].shift(1)
deltaYGDP_lag2 = data['deltaGDP'].shift(2) / data['GDP(current $USD)'].shift(2)
deltaYGDP_lag3 = data['deltaGDP'].shift(3) / data['GDP(current $USD)'].shift(3)
GovtExp_lag1 = data['deltaGovtExp'].shift(1)
GovtExp_lag2 = data['deltaGovtExp'].shift(2)
GovtExp_lag3 = data['deltaGovtExp'].shift(3)
```

```python
X = pd.DataFrame({
    'deltaYGDP_lag1': deltaYGDP_lag1,
    'deltaYGDP_lag2': deltaYGDP_lag2,
    'deltaYGDP_lag3': deltaYGDP_lag3,
    'GovtExp_lag1': GovtExp_lag1,
    'GovtExp_lag2': GovtExp_lag2,
    'GovtExp_lag3': GovtExp_lag3,
    'GShock' : data['GShock']
})
X
```

```python
X = X.fillna(method='bfill')
```
