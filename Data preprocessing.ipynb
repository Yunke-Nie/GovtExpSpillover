```python
import pandas as pd
import numpy as np
```

```python
exchange_rate = pd.read_csv("./data/Official exchange rate, average of a year (Current US$).csv")
exchange_rate.head(3)
```

```python
exchange_rate.columns
```

```python
exchange_rate_df = pd.melt(exchange_rate, id_vars=['countrycode', 'countryname', 'region', 'regionname', 'adminregion',
       'adminregionname', 'incomelevel', 'incomelevelname', 'lendingtype',
       'lendingtypename', 'indicatorname', 'indicatorcode'], var_name="year", value_name="value")
```

```python
exchange_rate_df.head(2)
```

```python
exchange_rate_df = exchange_rate_df.drop(['region', 'regionname', 'adminregion',
       'adminregionname', 'incomelevel', 'incomelevelname', 'lendingtype',
       'lendingtypename', 'indicatorname', 'indicatorcode'], axis=1)
```

```python
exchange_rate_df.head(10)
```

```python
exchange_rate_df.to_csv('./data/cleaned/exchange_rates.csv', index=False)
```
